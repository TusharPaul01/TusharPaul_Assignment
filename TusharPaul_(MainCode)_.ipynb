```python
# Load the model and tokenizer
model = AutoModelForCausalLM.from_pretrained("microsoft/phi-2", torch_dtype="auto", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2", trust_remote_code=True)
```

```python
def generate_response(tex, cache={},max_lines=3):
    if tex not in cache:
        inputs = tokenizer(tex, return_tensors="pt", return_attention_mask=False)

        # Generate text with the model
        outputs = model.generate(**inputs, max_length=200, temperature=1, num_return_sequences=1)
        text = tokenizer.batch_decode(outputs)[0]

        # Cache the result
        cache[tex] = text

        # Split the text into lines
        lines = text.split('\n')

        # Limit the response to max_lines
        return lines[:max_lines]
    else:
        # Split the cached text into lines
        return cache[tex].split('\n')
```

```python
while True:
    # Get input from the user
    tex = input("Ask a question or type 'exit' to end: ")

    # Check if the user wants to exit
    if tex.lower() == 'exit':
        print("Goodbye!")
        break

    # Generate response
    response = generate_response(tex, max_lines=3)

    # Print the first three lines of the response
    lines = '\n'.join(response[:2])
    print(lines)
    print()
```
